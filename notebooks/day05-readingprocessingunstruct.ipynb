# Reading + Processing Unstructured Data

With sample documents in data/synthetic_samples

```python
import pdfplumber as pdf
from pathlib import Path

DPATH = Path(r"C:\\Users\\micro\\Documents\\ABTalksAI-Cohort\\data\\synthetic_samples\\")
PATH = Path(r"C:\\Users\\micro\\Documents\\ABTalksAI-Cohort\\")
OUTPUT = PATH / "raw_text"
OUTPUT.mkdir(parents=True, exist_ok=True)

output_text = ""

with pdf.open(DPATH / "summary_of_benefits_and_coverage.pdf") as pdf:
    for page_number, page in enumerate(pdf.pages, start=1):
        text = page.extract_text() or ""

        output_text += f"\n--- Page {page_number} ---\n\n"
        output_text += text
        output_text += "\n"

(OUTPUT / "benefits.txt").write_text(
    output_text,
    encoding="utf-8"
)
print("success.")
```

```python
from docx import Document

document = Document(DPATH / "claims_process_guide.docx")

output_text = ""

for paragraph in document.paragraphs:
    text = paragraph.text.strip()
    if text:
        output_text += text + "\n"

# for simplicity we ignore tables

(OUTPUT / "claims_process.txt").write_text(
    output_text,
    encoding = "utf-8"
)

print("success")
```

```python
# Tesseract on PATH
import pytesseract
from pdf2image import convert_from_path

image_path = DPATH / "scanned_enrollment_form.pdf"

page_images = convert_from_path(
    image_path,
    dpi=300
)

output_text = ""

for page_number, page_image in enumerate(page_images, start=1):
    text = pytesseract.image_to_string(
        page_image,
        lang="eng"
    )
    output_text += f"\n--- Page {page_number} ---\n\n"
    output_text += text.strip()

(OUTPUT / "enrollment.txt").write_text(
    output_text,
    encoding="utf-8"
)
print("success.")
```

### Website FAQ Scrape

AI found me this website:
https://www.healthcare.gov/using-marketplace-coverage/getting-medical-care/?utm_source=chatgpt.com

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.healthcare.gov/using-marketplace-coverage/getting-medical-care/?utm_source=chatgpt.com"

response = requests.get(
    url,
    headers={
        "User-Agent": "Mozilla/5.0"
    },
    timeout=30
)

response.raise_for_status()
soup = BeautifulSoup(response.text, "html.parser")
main = soup.find("main")
text = main.get_text("\n", strip=True)

(OUTPUT / "medical_care_FAQ.txt").write_text(
    text,
    encoding="utf-8"
)
print("success.")
```
