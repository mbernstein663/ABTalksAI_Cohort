## Collections

ChromaDB stores information inside containers called **collections**. Each collection stores a set of information for each chunk, including text, embedding, and metadata- in objects called **records**. These are affected using ordinary CRUD operations such as:

```
.add()
.query()
.get()
.update()
.delete()
```

In our use case, each chunk will get its own record inside a collection.

````python
# calling the collection

coverage_collection = client.get_collection(
    name="coverage_kb"
)

print("Confirmed:", coverage_collection.name)
````
